```
years = list(range(2015, 2021))
season_recaps = []
ind_team_vegas = []
```

```
for i in years:
  szn_url = 'https://www.pro-football-reference.com/years/' + str(i) + '/'
  szn = pd.read_html(szn_url)
  szn = pd.concat(szn)
  szn['Year'] = str(i)
  season_recaps.append(szn)
  print('Scraped: ' + str(i) + ' season recap')
  for p in names_dict.keys():
    for c in names_dict[p]:
      odds_url = 'https://www.pro-football-reference.com/teams/' + names_dict[p][c]['scrap_abb'] + '/' + str(i) + '_lines.htm'
      tm_odds = pd.read_html(odds_url)
      tm_odds = tm_odds[0]
      tm_odds['Year'] = str(i)
      tm_odds['Team_abb'] = names_dict[p][c]['scrap_abb']
      tm_odds['Team_Full'] = names_dict[p][c]['team_name']
      if 'old_name' in names_dict[p][c]:
        tm_odds['Team_Full'] = names_dict[p][c]['old_name']
      print('Scraped: ' + names_dict[p][c]['scrap_abb'] + str(i) + 'season odds')
      ind_team_vegas.append(tm_odds)
      print('Appended: ' + names_dict[p][c]['scrap_abb'] + str(i) + 'season odds')
```

```
def rowRemover(df):
  # Goal: remove every 5 rows
  #df.reset_index(inplace=True)
  df = df.iloc[lambda x: x.index % 5 != 0]
  return df
```

```
season_recaps = [rowRemover(x) for x in season_recaps]
```

```
def editTmNames(df):
  import re
  df['Tm'] = df['Tm'].str.replace('[^a-zA-Z\s]', '')
  return df
```

```
season_recaps = [editTmNames(x) for x in season_recaps]
```

```
def addHomeTeam(df):
  df['home_team'] = np.where(df['Opp'].str.contains('@'), df.Opp, df.Team_Full)
  return df
```

```
ind_team_vegas = [addHomeTeam(x) for x in ind_team_vegas]
```

```
def removeAtSymbol(df):
  import re
  df['Opp'] = df['Opp'].str.replace('[^a-zA-Z\s]', '')
  df['home_team'] = df['home_team'].str.replace('[^a-zA-Z\s]', '')
  return df
```

```
ind_team_vegas = [removeAtSymbol(x) for x in ind_team_vegas]
```

```
def convertTeamAbbs(df, da_dict):
  tst_lst_name = list(range(0, len(df)))
  tst_lst_current = list(range(0, len(df)))
  for v in range(0, len(df)):
    for i in da_dict.keys():
      for p in da_dict[i]:
        #for v in range(0, len(df)):
        if df['Opp'][v] == da_dict[i][p]['odds_pg_abb']:
          tst_lst_name[v] = da_dict[i][p]['team_name']
          tst_lst_current[v] = da_dict[i][p]['team_name']
        if 'old_name' in da_dict[i][p].keys():
          if df['Opp'][v] == da_dict[i][p]['old_abb']:
            tst_lst_name[v] = da_dict[i][p]['old_name']
            tst_lst_current[v] = da_dict[i][p]['team_name']
  df['Opp_Full'] = tst_lst_name
  df['Opp_Full_Current'] = tst_lst_current
  return df
```

```
convertTeamAbbs(tst, names_dict)
```

```
ind_team_vegas = [convertTeamAbbs(x, names_dict) for x in ind_team_vegas]
```

```
def finishHomeTeam(df):
  for i in range(0, len(df['home_team'])):
    if len(df['home_team'][i]) < 4:
      df['home_team'][i] = df['Opp_Full_Current'][i]
  return df
```

```
ind_team_vegas = [finishHomeTeam(x) for x in ind_team_vegas]
```
